```python
import pandas as pd
import numpy as np
import os

np.random.seed(42)

data_folder = r'C:\Users\user\Desktop\ecommerce-analytics-project\data'
os.makedirs(data_folder, exist_ok=True)

n = 5000

df = pd.DataFrame({
    'order_id': [f'ORD{str(i).zfill(5)}' for i in range(1, n+1)],
    'customer_id': [f'CUST{str(np.random.randint(1, 2000)).zfill(4)}' for _ in range(n)],
    'order_date': pd.date_range('2023-01-01', '2024-12-31', periods=n),
    'order_status': np.random.choice(
        ['Delivered', 'Shipped', 'Processing', 'Cancelled'],
        n, p=[0.85, 0.08, 0.05, 0.02]),
    'total_amount': np.random.uniform(20, 500, n).round(2),
    'payment_method': np.random.choice(
        ['Credit Card', 'Debit Card', 'PayPal', 'Cash'],
        n, p=[0.50, 0.25, 0.15, 0.10]),
    'product_category': np.random.choice(
        ['Electronics', 'Clothing', 'Home & Garden', 'Books', 'Sports', 'Beauty'], n),
    'customer_city': np.random.choice(
        ['Sao Paulo', 'Rio de Janeiro', 'Brasilia', 'Salvador', 'Fortaleza'], n),
    'customer_rating': np.random.choice(
        [1, 2, 3, 4, 5], n, p=[0.05, 0.05, 0.10, 0.30, 0.50]),
    'delivery_days': np.random.randint(1, 15, n)
})

df.to_csv(os.path.join(data_folder, 'ecommerce_data.csv'), index=False)

print("✅ Dataset created successfully!")
print(f"Total Orders: {len(df):,}")
print(f"Total Revenue: ${df['total_amount'].sum():,.2f}")
print(df.head())
```
